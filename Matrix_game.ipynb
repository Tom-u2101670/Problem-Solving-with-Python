```python
def change_strat(move_matrix,n):
    new_moves = np.copy(move_matrix)
    for j in range(n):
        if not (move_matrix[0,j] == 0):
            pos = True
            neg = True
            for k in range(n-1):
                if (move_matrix[k+1, ((j+k+1) %n)] == 0):
                    pos = False
                if (move_matrix[k+1, ((j-k-1) %n)] == 0):
                    neg = False
                if (pos == False) and (neg == False):
                    break
            if (pos == True):
                for q in range(n):
                    if not (new_moves[q,(j+q)%n] == 20):
                        new_moves[q,(j+q)%n] = 1
            if (neg == True):
                for q in range(n):
                    if not (new_moves[q,(j-q)%n] == 20):
                        new_moves[q,(j-q)%n] = 1
    indices = np.where(new_moves == 2)
    l = len(indices[0])
    for r in range(l):
        ind_i = indices[0][r]
        ind_j = indices[1][r]
        new_moves[ind_i, ind_j] = 0
    if np.any(new_moves == 1):
        return new_moves, True
    else:
        return new_moves, False
```
